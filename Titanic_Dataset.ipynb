Titanic Survival Analysis

1. Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Load dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

Preview the data


```python
print(df.head())
print(df.info())
```

Task 1: Average age of survivors vs non-survivors


```python
avg_age = df.groupby("Survived")["Age"].mean()
print("\nAverage Age of Survivors vs Non-Survivors:")
print(avg_age)
```

Task 2: Survival rate by passenger class


```python
survival_rate = df.groupby("Pclass")["Survived"].mean() * 100
print("\nSurvival Rate by Passenger Class (%):")
print(survival_rate)
```

Task 3: Plot bar chart for gender survival


```python
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=df, ci=None)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

Task 4: Write down 3 insights (printed as text)


```python
print("\n--- Insights ---")
print("1. Survivors tend to be younger than non-survivors.")
print("2. First-class passengers had a much higher survival rate than second and third class.")
print("3. Women had a significantly higher chance of survival compared to men.")
```
